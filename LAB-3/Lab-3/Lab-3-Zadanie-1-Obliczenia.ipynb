```python
import pandas as pd
from enum import StrEnum
import importlib

class SilaWiatru(StrEnum):
    silny = 'silny'
    slaby = 'słaby'
    brak = 'brak'

class Zachmurzenie(StrEnum):
    pochmurnie = 'pochmurnie'
    slonecznie = 'słonecznie'

class Temperatura(StrEnum):
    zimno = 'zimno'
    cieplo = 'ciepło'
    goraco = 'gorąco'
    
class Zagrano(StrEnum):
    tak = 'tak'
    nie = 'nie'

kolumny_x = [
    "Siła wiatru",
    "Zachmurzenie",
    "Odczuwalna temperatura"
]

kolumna_y = 'Zagrano mecz'


dane_uczace = pd.DataFrame({
    kolumny_x[0]: [
        SilaWiatru.silny,
        SilaWiatru.silny,
        SilaWiatru.brak,
        SilaWiatru.brak,
        SilaWiatru.slaby,
        SilaWiatru.slaby,
        SilaWiatru.brak,
        SilaWiatru.silny,
        SilaWiatru.brak,
        SilaWiatru.silny,
    ],
    kolumny_x[1]: [
        Zachmurzenie.pochmurnie,
        Zachmurzenie.pochmurnie,
        Zachmurzenie.slonecznie,
        Zachmurzenie.slonecznie,
        Zachmurzenie.pochmurnie,
        Zachmurzenie.slonecznie,
        Zachmurzenie.pochmurnie,
        Zachmurzenie.slonecznie,
        Zachmurzenie.pochmurnie,
        Zachmurzenie.pochmurnie,
    ],
    kolumny_x[2]: [
        Temperatura.zimno,
        Temperatura.cieplo,
        Temperatura.cieplo,
        Temperatura.goraco,
        Temperatura.goraco,
        Temperatura.cieplo,
        Temperatura.zimno,
        Temperatura.zimno,
        Temperatura.goraco,
        Temperatura.cieplo
    ],
    kolumna_y: [
        Zagrano.nie,
        Zagrano.nie,
        Zagrano.tak,
        Zagrano.nie,
        Zagrano.tak,
        Zagrano.tak,
        Zagrano.nie,
        Zagrano.tak,
        Zagrano.tak,
        Zagrano.tak
    ]
})

dane_testowe = pd.DataFrame({
    kolumny_x[0]: [SilaWiatru.silny],
    kolumny_x[1]: [Zachmurzenie.slonecznie],
    kolumny_x[2]: [Temperatura.cieplo],
})
```

```python
klasyfikatory = importlib.import_module('Lab-3-Zadanie-1-Klasyfikatory')

X_train = dane_uczace.drop(kolumna_y, axis=1)
y_train = dane_uczace[kolumna_y]
X_test = dane_testowe

result_bayes = klasyfikatory.naive_bayes(X_train, y_train, X_test)

print("Naiwny klasyfikator Bayesa")
klasyfikatory.print_result(X_test, result_bayes, kolumna_y)
```

```python
klasyfikatory = importlib.import_module('Lab-3-Zadanie-1-Klasyfikatory')

X_train = dane_uczace.drop(kolumna_y, axis=1)
y_train = dane_uczace[kolumna_y]
X_test = dane_testowe

for k in range(1, 5):
    print(f"\n\nk = {k}")
    result_knn_euler = klasyfikatory.knn(X_train, y_train, X_test, k, klasyfikatory.euler_dist)
    result_knn_manhattan = klasyfikatory.knn(X_train, y_train, X_test, k, klasyfikatory.manhattan_dist)

    print("Klasyfikator kNN - odległość Euklidesowa")
    klasyfikatory.print_result(X_test, result_knn_euler, kolumna_y)

    print("\nKlasyfikator kNN - odległość miejska")
    klasyfikatory.print_result(X_test, result_knn_manhattan, kolumna_y)
```
