```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("UsersQ2023.csv", delimiter=";")
df.head()
df.tail()
```

```python
new_rows = pd.DataFrame([{"NoQ": np.nan, "Q": "Q1 '08"}, {"NoQ": np.nan, "Q": "Q2 '08"}])
```

```python
df = pd.concat([new_rows, df], ignore_index=True)
```

```python
df.head()
```

```python
df["NoQ"] = range(1, len(df) + 1)
```

```python
df.head()
```

```python
df.to_csv("UsersQ2023_cleaned.csv")
```

```python
df.head()
```

```python
df_2 = df[df["Users"].isna() == False].copy()
df_2.loc[:, "Q"] = df_2["Q"].str.replace("'", "")
df_2.loc[:, "Q"] = df_2["Q"].str.replace(" ", "")
```

```python
df_2.head()
```

```python
df_2.dtypes
```

```python
df_2.loc[:, "Q"] = df_2["Q"].str[2:]
```

```python
df_2.head()
```

```python
df_2.tail()
```

```python
df_2.drop("NoQ", axis="columns", inplace=True)
grouped_users_per_q = df_2.groupby("Q").sum("Users")
```

```python
grouped_users_per_q.head()
len(grouped_users_per_q)
```

```python
grouped_users_per_q.to_csv("UserCountPerQ.csv")
```

```python
revenue_df = pd.read_csv("RevenueByYear.csv", delimiter=";")
```

```python
revenue_df.loc[:, "Year"] = revenue_df["Year"].astype(str)
revenue_df.loc[:, "Year"] = revenue_df["Year"].str[2:]
# grouped_users_per_q
revenue_df
```

```python
len(revenue_df) == len(grouped_users_per_q)
```

```python
revenue_df.set_index(["Year"])
```

```python
revenue_to_users_df = revenue_df.merge(grouped_users_per_q, left_on="Year", right_on="Q")
```

```python
revenue_to_users_df.head()
```

```python
revenue_to_users_df.to_csv("RevenueToUsersByQ.csv")
```
