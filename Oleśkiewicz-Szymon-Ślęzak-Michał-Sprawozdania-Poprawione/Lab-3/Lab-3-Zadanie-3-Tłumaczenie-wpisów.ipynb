Tłumaczenie z wykorzystaniem LibreTranslate działającego na maszynie wirtualnej hosta w porcie 5000

https://libretranslate.com/

```python
import requests
from json import loads
import csv

URL = "http://localhost:5000/translate"

with open('Lab-3-Zadanie-3-Dane-Mandrill.csv', 'r') as f:
    reader = csv.reader(f)
    mandrill_data = list(x for x in reader if x != [])
    mandrill_data.pop(0)

with open('Lab-3-Zadanie-3-Dane-Mandrill-Przetłumaczone.csv', 'w') as f:
    f.write("Post\n")

for line in mandrill_data:
    body = {
        "q": line,
        "source": "auto",
        "target": "en",
        "format": "text",
        "alternatives": 3,
        "api_key": ""
    }
    r = requests.post(URL, json=body)
    j = loads(r.text)

    translated_text = j['translatedText']
    with open('Lab-3-Zadanie-3-Dane-Mandrill-Przetłumaczone.csv', 'a') as f:
        f.write(" ".join(translated_text)+'\n')
```

```python
import requests
from json import loads
import csv

URL = "http://localhost:5000/translate"
with open('Lab-3-Zadanie-3-Dane-Inne.csv', 'r') as f:
    reader = csv.reader(f)
    other_data = list(x for x in reader if x != [])
    other_data.pop(0)

with open('Lab-3-Zadanie-3-Dane-Inne-Przetłumaczone.csv', 'w') as f:
    f.write("Post\n")

for line in other_data[97:]:
    body = {
        "q": line,
        "source": "auto",
        "target": "en",
        "format": "text",
        "alternatives": 3,
        "api_key": ""
    }
    r = requests.post(URL, json=body)
    j = loads(r.text)

    translated_text = j['translatedText']
    with open('Lab-3-Zadanie-3-Dane-Inne-Przetłumaczone.csv', 'a') as f:
        try:
            f.write(" ".join(translated_text) + '\n')
        except UnicodeEncodeError:
            pass
```
